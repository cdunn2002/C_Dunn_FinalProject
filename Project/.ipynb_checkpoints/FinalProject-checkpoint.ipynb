```python
import pandas as pd
import numpy as np
import netCDF4 as nc
import xarray as xr #USE XARRAY TO OPEN .NC FILES
import matplotlib.pyplot as plt
# .where
```

```python
filepath = 'Final.Project/deployment0001_CE01ISSP-SP001-06-NUTNRJ000-recovered_cspp-nutnr_j_cspp_instrument_recovered_20140417T222823.449000-20140514T070004.817000.nc'
data_nitrate = xr.open_dataset(filepath)
display(data_nitrate)
```

```python
data = data_nitrate.where(data_nitrate.time.dt.month == 4, drop=True)
```

```python
# Getting values for depth and nitrate
nitrate = data.nitrate_concentration.values
depth = data.depth.values
```

```python
# Find average nitrate per depth
rounded_depth = np.round(depth)
var = pd.DataFrame(data = {'Depth': rounded_depth, 'Nitrate': nitrate}, columns = ['Depth','Nitrate'])
var = var.groupby(['Depth']).mean()
display(var)
```

```python
# Plot nitrate and oxygen 
# Make figure
fig,axs = plt.subplots(figsize=(8,8))
 

# Plot temperature and pressure
x = var.Nitrate
y = var.index
axs.scatter(x, y, color='k', lw=2, marker='o')


# Format figure
axs.invert_yaxis()
axs.grid()
# make a title and set axes labels, maybe increase font size
axs.set_title('Nitrate vs. Depth profile')
axs.set_xlabel('Nitrate (uMol L-1)', fontsize=16)
axs.set_ylabel('Depth (m)', fontsize=16)
axs.plot()
plt.savefig('figure1ocean215.png')
```

```python
# Phytoplankton Growth
filepath = '2010sConditionsNopHResponse.nc'
data_phyto = xr.open_dataset(filepath)
display(data_phyto)
```

```python
# Variable I want 'PicoPhytopl_N'
picophytopl_data = data_phyto['PicoPhytopl_N'].values

oxygen_data = data_phyto['Oxygen'].values
```

```python
# Getting dimensions of the data set
print("Dimensions of picophytopl_data:", np.ndim(picophytopl_data))

flatphyto_data = picophytopl_data.flatten()
```

```python
# Rounding our values for a more easily digestable graph
rounded_oxygen = np.round(oxygen_data, 1)
rounded_picophytopl_data = np.round(picophytopl_data, 1)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# 3D graph for Phytoplankton (3 dimensional data)
picophytopl_data = data_phyto['PicoPhytopl_N'].values

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 

# Plotting the phytoplankton data
ax.scatter(picophytopl_data[0], picophytopl_data[1], picophytopl_data[2])
ax.set_title('Phytoplankton Data')
x_max = 2.5
x_min = 0.0
ax.set_xlim([x_min, x_max])
ax.set_xlabel('Biomass (mg N m^-3)')
ax.set_ylabel('Length (m)')
z_max = 20
z_min = 0
ax.set_zlim([z_max, z_min])
ax.set_zlabel('Depth (m)')

# Show the plot
plt.show()
plt.savefig('figure2ocean215.png')
```
